# Data Summary
This file displays a general summary of the data.
- Comparison on how many cases there is Bankruptcy or not


```python
from ucimlrepo import fetch_ucirepo 
  
# Downlaod the Dataset 
taiwanese_bankruptcy_prediction = fetch_ucirepo(id=572) 
  
# Get the features and target variables
X = taiwanese_bankruptcy_prediction.data.features 
y = taiwanese_bankruptcy_prediction.data.targets 
```

## Bankruptcy vs Non-Bankruptcy 

```python
import matplotlib.pyplot as plt

# Count the occurrence of each category (bankruptcy and non-bankruptcy)
counts = y['Bankrupt?'].value_counts()

# DIsplay the Pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=counts.index.map({0: 'Non-Bankruptcy', 1: 'Bankruptcy'}).astype(str), autopct='%1.1f%%', startangle=25)
plt.title('Proportion of Bankruptcies to Non-Bankruptcies')
plt.axis('equal')  

plt.show()
```
